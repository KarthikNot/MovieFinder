```python
# Text Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

vectors = tfidf.fit_transform(df['all_tags'])
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity

top_k = 10
row_idx = df.index[df['id'] == 862][0]
sim = cosine_similarity(vectors[row_idx], vectors).flatten()
top_indices = np.argsort(sim)[-top_k-1:-1][::-1]

df.iloc[top_indices]['title']
```
